#**FIR Filter Designing**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import scipy.fft
from scipy.signal import filtfilt
from numpy import sin,cos,pi,exp,sqrt
from scipy.signal import butter, lfilter
from scipy import signal
from scipy.io import wavfile, loadmat
import scipy
import time

```

```python
w_c = 0.65 * np.pi
N = 4
T = (N-1)/2
h_n = []
for n in range(N):
  h_n.append(np.sin(w_c*(n-T))/(np.pi*(n-T)))
h_n.insert(2,(np.cos(w_c*(T-T))*w_c)/(np.pi))
# h_n.append(sin(w_c*(T-T))/(np.pi*(4-T)))
print(h_n)
```

```python
plt.plot(h_n)
```

```python
#hamming window:-
w_n = []
for i in range(N+1):
  print(i)
  w_n.append(0.54 - 0.46*np.cos((2*np.pi*i)/(N)))
print("w_n is:",w_n)

h_d1 = []
for i in range(N+1):
  h_d1.append(h_n[i] * w_n[i])
  print(h_d1)
```

```python
plt.plot(h_d1)
```

```python
#rectangular:-
r_n = [1,1,1,1,1]
h_d2 = []
for i in range(N+1):
  h_d2.append(h_n[i] * r_n[i])
  print(h_d2)
```

```python
print(h_d2)
plt.plot(h_d2)
```

```python
#bertlett window:-
b_t = []
mid = (N+1)/2
for i in range(N+1):
  if (i < mid):
    b_t.append(2*i/(N))
  else:
    b_t.append(2 - 2*i/(N))

h_d3 = []
for i in range(N+1):
  h_d3.append(h_n[i] * b_t[i])
  print(h_d3)
```

```python
N = 300
fs = 10000
t = (1/fs)*np.arange(0,N)
f1 = 2000
f2 = 4000
x = sin(2*pi*f1*t)+sin(2*pi*f2*t)

plt.plot(t,x)
plt.title('X(t) signal')
plt.xlabel('t (in sec)')
plt.grid()
```

```python
X = np.fft.fft(x)
H_d1 = np.fft.fft(h_d1,300)
Y1 = X * H_d1
plt.plot(Y1)
```

```python
#using hamming window:-

_, ax = plt.subplots(3,1, figsize = (15, 9), constrained_layout = True)
y1 = scipy.ifft(Y1)
ax[0].plot(t,x)
ax[1].plot(h_d1)
ax[1].stem(h_d1)
ax[2].plot(y1)
```

```python
H_d2 = np.fft.fft(h_d2,300)
Y2 = X * H_d2
y2 = scipy.ifft(Y2)
```

```python
#using rectangular window:-
_, ax = plt.subplots(3,1, figsize = (15, 9), constrained_layout = True)
ax[0].plot(t,x)
ax[1].plot(h_d2)
ax[1].stem(h_d2)
ax[2].plot(y2)
```

```python
H_d3 = np.fft.fft(h_d3,300)
Y3 = X * H_d3
y3 = scipy.ifft(Y3)
```

```python
#using bartlett window:-
_, ax = plt.subplots(3,1, figsize = (15, 9), constrained_layout = True)
ax[0].plot(t,x)
ax[1].plot(h_d3)
ax[1].stem(h_d3)
ax[2].plot(y3)
```
